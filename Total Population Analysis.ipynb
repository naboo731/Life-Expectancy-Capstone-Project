```python
import numpy as np
import pandas as pd
import plotly
import matplotlib.pyplot as plt
from scipy import stats
import plotly.figure_factory as ff
import pycountry
import geopandas as geopandas
import folium
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

init_notebook_mode(connected=True)
```

```python
data = pd.read_csv('total_pop.csv', skiprows=3)
data.drop(inplace = True, columns=['Indicator Name', 'Indicator Code', '2020', 'Unnamed: 65'])
data
```

```python
data.dropna(inplace=True, thresh=len(data.columns)-45)
data
```

```python
data = data.interpolate(method ='linear', limit_direction ='forward')
data
```

```python
world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))
world.columns=['pop_est', 'continent', 'name', 'Country Code', 'gdp_md_est', 'geometry']

merged_data = pd.merge(world, data, on = 'Country Code')

location = pd.read_csv('https://raw.githubusercontent.com/melanieshi0120/COVID-19_global_time_series_panel_data/master/data/countries_latitude_longitude.csv')
merged_data = merged_data.merge(location, on = 'name').reset_index()
merged_data.drop(inplace=True, columns=['Country Name'])
merged_data.head()
```

```python
merged_data.plot(column='pop_est', scheme='quantiles', figsize=(25,20), legend=True, cmap='coolwarm')
plt.title('Estimated Population in Different Countries', fontsize=25)

plt.show()
```

```python
merged_data.plot(column='1960', scheme='quantiles', figsize=(25,20), legend=True, cmap='coolwarm')
plt.title('Life Expectancy in 1960 in Different Countries', fontsize=25)

plt.show()
```

```python
map1 = folium.Map()
folium.Choropleth(
    geo_data = merged_data,
    name = 'choropleth',
    data = merged_data,
    columns=['name', '1960'],
    key_on = 'feature.properties.name',
    fill_color='OrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Average Life Expectancy in Years'
).add_to(map1)

map1
```

```python
new_df = merged_data.drop(columns=['pop_est', 'Country Code', 'gdp_md_est', 'geometry'])
fixed_df = (new_df.set_index(['name', 'longitude', 'latitude', 'continent'])
                    .stack()
                    .rename_axis(['name', 'longitude', 'latitude','continent', 'year'])
                    .reset_index(name='Avg Life Expectancy'))
fixed_df.drop(index=0, inplace=True)

fixed_df
```

```python
slider = []
scl = [[0.0, '#ffffff'],[0.2, '#b4a8ce'],[0.4, '#8573a9'],
       [0.6, '#7159a3'],[0.8, '#5732a1'],[1.0, '#2c0579']]

for year in fixed_df.year.unique():
    df_sectioned = fixed_df[fixed_df['year'] == year]
    
    data_one_year = dict(
                        type='choropleth',
                        locations = fixed_df['name'],
                        z=fixed_df['Avg Life Expectancy'].astype(float),
                        locationmode='country names',
                        colorscale = scl
                        )
    slider.append(data_one_year)
```

```python
steps = []

for i in range(len(slider)):
    step = dict(method='restyle',
               args=['visible', [False] * len(slider)],
               label='Year {}'.format(i + 1960))
    step['args'][1][i] = True
    steps.append(step)
    
sliders = [dict(active=0, pad={'t':1}, steps=steps)]
```

```python
layout = dict(geo=dict(scope='world',
                       projection={'type': 'natural earth'}),
              sliders=sliders)
fig = dict(data=slider, layout=layout)
plotly.offline.iplot(fig)
```
