```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import geopandas
```

```python
female_data = pd.read_csv('females.csv', skiprows =3)
female_data.drop(columns=['Indicator Name', 'Indicator Code', '2020', 'Unnamed: 65'], inplace=True)
female_data.head()
```

```python
female_countries = female_data.groupby('Country Name', as_index=False)['2019'].mean()
female_countries.sort_values(by = '2019', ascending = False).head(10)
```

```python
female_countries = female_data.groupby('Country Name', as_index=False)['2019'].mean()
female_countries.sort_values(by = '2019', ascending = True).head(10)
```

```python
aruba = female_data.iloc[0]['1960':'2019']
aruba.plot.line()

plt.title('Aruba', y=1.03)
plt.xlabel('Recorded Year')
plt.ylabel('Life Expectancy (years)')
```

```python
es_africa = female_data.iloc[1]['1960':'2019']
es_africa.plot.line(color='orange')

plt.title('Africa Eastern and Southern', y=1.03)
plt.xlabel('Recorded Year')
plt.ylabel('Life Expectancy (years)')
```

```python
female_data.isnull().sum()
```

```python
female_data.dropna(inplace=True, thresh=len(female_data.columns)-45)
```

```python
female_data = female_data.interpolate(method ='linear', limit_direction ='forward')
female_data.head(10)
```

```python
female_data.isnull().sum()
```

```python
world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))
world.columns=['pop_est', 'continent', 'name', 'Country Code', 'gdp_md_est', 'geometry']

full_female_data = pd.merge(world, female_data, on = 'Country Code')

location = pd.read_csv('https://raw.githubusercontent.com/melanieshi0120/COVID-19_global_time_series_panel_data/master/data/countries_latitude_longitude.csv')
complete_data = full_female_data.merge(location, on = 'name').reset_index()
complete_data.drop(inplace=True, columns=['Country Name'])
complete_data.head()
```

```python
new_df = complete_data.drop(columns=['Country Code', 'gdp_md_est', 'geometry'])
fixed_df = (new_df.set_index(['name', 'longitude', 'latitude', 'continent', 'pop_est'])
                    .stack()
                    .rename_axis(['name', 'longitude', 'latitude', 'continent', 'pop_est', 'year'])
                    .reset_index(name='Avg Life Expectancy'))
fixed_df.drop(index=0, inplace=True)
fixed_df.drop(fixed_df.loc[fixed_df.year == 'index'].index, inplace=True)
fixed_df
```

```python
fixed_df.to_csv('complete_female.csv')
```
