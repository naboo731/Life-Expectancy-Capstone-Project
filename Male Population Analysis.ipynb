```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
male_data = pd.read_csv('males.csv', skiprows=3)
male_data.drop(inplace=True, columns=['Indicator Name', 'Indicator Code', '2020', 'Unnamed: 65'])
male_data.head()
```

```python
top_male_countries = male_data.groupby('Country Name', as_index=False)['2019'].mean()
top_male_countries.sort_values(by='2019', ascending=False).head(10)
```

```python
bottom_male_countries = male_data.groupby('Country Name', as_index=False)['2019'].mean()
bottom_male_countries.sort_values(by='2019', ascending=True).head(10)
```

```python
usa = male_data.iloc[251]['1960':'2019']
usa.plot.line()

plt.title('United States', y=1.03)
plt.xlabel('Recorded Year')
plt.ylabel('Life Expectancy (years)')
```

```python
swiss = male_data.iloc[235]['1960':'2019']
swiss.plot.line()

plt.title('Switzerland', y=1.03)
plt.xlabel('Recorded Year')
plt.ylabel('Life Expectancy (years)')
```
